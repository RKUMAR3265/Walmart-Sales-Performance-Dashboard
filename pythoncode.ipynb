```python
import pandas as pd
df = pd.read_csv(r"D:\DATA ANALYSIS\walmart sales\Walmart_SQL_READY_FINAL_CLEAN.csv")
print(df.head())
print(df.columns)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
num_cols = [
    'sales', 'profit', 'discount', 'unit_price',
    'order_quantity', 'shipping_cost',
    'product_base_margin', 'customer_age'
]
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df['customer_age'] = df['customer_age'].fillna(df['customer_age'].median())
```

```python
df['customer_age'].isnull().sum()
```

```python
df.groupby(df['order_date'].dt.to_period('M'))['sales'].sum()
```

```python
import numpy as np
df['customer_age'] = np.random.randint(18, 65, size=len(df))
```

```python
df.loc[df['customer_segment'] == 'Consumer', 'customer_age'] = \
    df.loc[df['customer_segment'] == 'Consumer', 'customer_age'].fillna(30)
df.loc[df['customer_segment'] == 'Corporate', 'customer_age'] = \
    df.loc[df['customer_segment'] == 'Corporate', 'customer_age'].fillna(40)
df.loc[df['customer_segment'] == 'Small Business', 'customer_age'] = \
    df.loc[df['customer_segment'] == 'Small Business', 'customer_age'].fillna(35)
```

```python
df[['customer_age', 'customer_segment']].head()
df['customer_age'].isnull().sum()
```

```python
df.to_csv(
    r"D:\DATA ANALYSIS\walmart sales\Walmart_SQL_READY_FINAL_CLEAN_UPDATED.csv",
    index=False
)
print("File saved successfully")
```
